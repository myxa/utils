```python
import nibabel as nib
import numpy as np
import glob
import os

nii_files = sorted(glob.glob(r"C:\Users\administrator\Desktop\howto\ROI_sets\bnta\VOIs\Subject_0001/sSST_01-0401-*.nii"))

imgs = [nib.load(f) for f in nii_files]
data_list = [img.get_fdata() for img in imgs]


```

```python
for i in range(len(data_list)):
    if data_list[i].shape != data_list[0].shape:
        print(i)
```

```python
# Stack along 4th dimension
four_d_data = np.stack(data_list[:-1], axis=-1)
affine = imgs[0].affine  # assume all identical
header = imgs[0].header
header.set_data_shape(four_d_data.shape)

four_d_img = nib.Nifti1Image(four_d_data, affine, header)
nib.save(four_d_img, "4d.nii.gz")
```
