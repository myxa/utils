```python
import nibabel as nib
import numpy as np
import glob
import os
```

```python
nii_files = sorted(glob.glob(
    r"C:\Users\administrator\Desktop\SST\sub-001\func\2\*.nii"))

imgs = [nib.load(f) for f in nii_files]
data_list = [img.get_fdata() for img in imgs]
```

```python
for i in range(len(data_list)):
    if data_list[i].shape != data_list[0].shape:
        print(i)
        to_del = i
```

```python
for i in range(3, 6):
    os.mkdir(rf'C:\Users\administrator\Desktop\SST\sub-001\func\{i}')
```

```python
del data_list[to_del]
```

```python
# Stack along 4th dimension
four_d_data = np.stack(data_list, axis=-1)
affine = imgs[0].affine  # assume all identical
header = imgs[0].header
header.set_data_shape(four_d_data.shape)

four_d_img = nib.Nifti1Image(four_d_data, affine, header)
```

```python
nib.save(four_d_img,
        r"C:\Users\administrator\Desktop\SST\sub-001\func\sub-001_task-sst_run-2_bold.nii.gz")
```
