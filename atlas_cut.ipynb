```python
import os
import nibabel as nib
import numpy as np

atlas_path = "C:\\Users\\administrator\\Desktop\\BN_Atlas_246_3mm.nii"          # your atlas
out_dir = "C:\\Users\\administrator\\Desktop\\BN_Atlas_246_3mm_ROIs"
os.makedirs(out_dir, exist_ok=True)

img = nib.load(atlas_path)
data = img.get_fdata().astype(int)
affine = img.affine
header = img.header


labels = np.unique(data)
labels = labels[labels != 0] 
```

```python
for lab in labels:
    mask = (data == lab).astype(np.uint8)
    out_img = nib.Nifti1Image(mask, affine, header)
    nib.save(out_img, os.path.join(out_dir, f"region_{lab:03d}.nii"))
```
